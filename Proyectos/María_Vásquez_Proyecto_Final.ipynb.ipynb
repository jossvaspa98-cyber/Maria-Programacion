Contiene datos de emisiones de di√≥xido de carbono (CO‚ÇÇ) de varios pa√≠ses entre 1975 y 2017.

üßæ Estructura de los datos
- Country: Pa√≠s

- Year: A√±o

- Type: Tipo de emisi√≥n:

    * Emisiones totales (kilotoneladas m√©tricas)

    * Emisiones per c√°pita (toneladas m√©tricas por persona)

- Value: Valor num√©rico de la emisi√≥n

üîç Tipos de emisiones
- Totales: CO‚ÇÇ emitido por todo el pa√≠s (kt)

- Per c√°pita: CO‚ÇÇ promedio por persona (t CO‚ÇÇ/persona)

    - Se calcula dividiendo emisiones totales por la poblaci√≥n

```python
#Importa el CSV como un DataFrame
import pandas as pd
import matplotlib.pyplot as plt

estimacion_emisiones_csv = "Estimaciones_Emisiones_dioxido_Carbono.csv"
df = pd.read_csv(estimacion_emisiones_csv, delimiter=',')
df
```

# Pregunta 0

1. Crear un dataframe con el nombre de variable 'df_emisions_co2' usando √∫nicamente las filas cuya columna 'Type' tiene el valor 'Emisiones totales de CO‚ÇÇ emitidas por el pa√≠s'

2. Crear un dataframe con el nombre de variable 'df_emisions_co2_per_capita' usando √∫nicamente las filas cuya columna 'Type' tiene el valor 'Emisiones promedio de CO‚ÇÇ emitidas por habitante'.

```python
#Crea un copia del original filtrado como df_emisions_co2

#Creamos una copia donde solo se insertan las filas donde el tipo sea "Emisiones (kilotoneladas metricas de dioxido de carbono)"
df_emisions_co2 = df[df["Type"] == "Emisiones (kilotoneladas metricas de dioxido de carbono)"].copy()

# Mostrar las primeras 5 filas del nuevo DataFrame para verificar el contenido
df_emisions_co2.head()
```

```python
#Crea un copia del original filtrado como df_emisions_co2_per_capita

#Creamos una copia donde solo se insertan las filas donde el tipo sea "Emisiones per capita (toneladas metricas de dioxido de carbono)"
df_emisions_co2_per_capita = df[df["Type"] == "Emisiones per capita (toneladas metricas de dioxido de carbono)"].copy()

# Mostrar las primeras 5 filas del nuevo DataFrame para verificar el contenido
df_emisions_co2_per_capita.head()
```

# Pregunta 4

Cu√°l es el pa√≠s que tiene mayor emisiones usando el dataframe 'df_emisions_co2'?

```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta

# Agrupamos los datos por pa√≠s Country y sumar las emisiones Value para cada uno
# Luego, se utiliza .idxmax() que devuelve el nombre del pa√≠s con la suma m√°s alta de emisiones
pais_con_mayor_emision = df_emisions_co2.groupby("Country")["Value"].sum().idxmax()

# Imprimir el resultado usando f para formatear el mensaje y que se vea mejor
print(f"El pa√≠s con mayores emsiones totales es: {pais_con_mayor_emision}")
```

# Pregunta 5

# Pregunta 8

Cu√°ntas son las emisiones en 'kilotoneladas m√©tricas' s√≥lo para Albania usando el dataframe 'df_emisions_co2'?

# Pregunta 9

Cu√°ntas son las emisiones totales en 'libras' s√≥lo para Costa Rica usando el dataframe 'df_emisions_co2'?


```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta
#1 tonelada m√©trica = 1.000 kilogramos
#1 kilogramo = 2.20462 libras

# Definimos el n√∫mero de libras que hay en un kilogramo
libras_por_kilogramo = 2.20462
# Calculamos cu√°ntos kilogramos hay en una kilotonelada m√©trica (1,000 * 1,000)
kilotonelada_metrica = 1000 * 1000

# Calculamos cu√°ntas libras hay en una kilotonelada m√©trica, parq realizar la conversi√≥n
conversion_libras = kilotonelada_metrica * libras_por_kilogramo

# Filtramos las emisiones de Costa Rica y convertimos el total a libras
total_emisiones_libras_cr = df_emisions_co2[df_emisions_co2["Country"] == "Costa Rica"]["Value"].sum() * conversion_libras

# Imprimir el resultado usando f para formatear el mensaje y que se vea mejor, resultado sin decimales
print(f"Las emisiones totales de Costa Rica en libras son: {total_emisiones_libras_cr:.0f}")
```

# Pregunta 10

Para s√≥lo 'Costa Rica' y suponiendo que las emisiones se estancan a partir del √∫ltimo a√±o reportado (se mantiene igual). Cu√°ntas son las emisiones en 'Kg' s√≥lo para Costa Rica desde el ultimo a√±o registrado hasta el 2025 usando el dataframe 'df_emisions_co2'?


```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta

# Filtramos el DataFrame para quedarnos solo con los datos de Costa Rica
df_cr = df_emisions_co2[df_emisions_co2["Country"] == "Costa Rica"]

# Obtenemos el √∫ltimo a√±o resgistrado
ultimo_anio_registrado = df_cr["Year"].max()

# Obtenemos el valor de emisiones del √∫ltimo a√±o registrado en kilotoneladas m√©tricas
emisiones_ultimo_anio = df_cr[df_cr["Year"] == ultimo_anio_registrado]["Value"].iloc[0]

# Calculamos cu√°ntos a√±os hay desde el √∫ltimo a√±o registrado hasta 2025
anios_proyectados = 2025 - ultimo_anio_registrado

# Multiplicamos las emisiones del √∫ltimo a√±o por la cantidad de a√±os proyectados
emisiones_totales_kilotoneladas = emisiones_ultimo_anio * anios_proyectados

# Convertimos las emisiones proyectadas a kilogramos (1 kilotonelada = 1000000 kg)
emisiones_proyectadas_kilogramos = emisiones_totales_kilotoneladas * kilotonelada_metrica

# Imprimir el resultado usando f para formatear el mensaje y que se vea mejor, resultado sin decimales
print(f"Emisiones proyectadas para Costa Rica desde {ultimo_anio_registrado + 1} hasta 2025 en kilogramos es: {emisiones_proyectadas_kilogramos:.0f}")
```

# Pregunta 12

Para s√≥lo los pa√≠ses de Centroam√©rica ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua'] obtener un reporte por a√±o de las emisiones totales en Centroam√©rica usando el dataframe 'df_emisions_co2'?
Se espera un dataframe de 2 columnas: 'Year' y 'total_por_a√±o'

```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta

#lista de paises de Centroam√©rica
paises_centroamerica = ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua']

#filtramos el dataframe para incluir solo los paises de centroam√©rica
#el .isin() funciona para filtrar filas que tienen un valor dentro de una lista espec√≠fica
df_centroamerica = df_emisions_co2[df_emisions_co2["Country"].isin(paises_centroamerica)]

#agrupamos por a√±o y sumar las emisiones de todos los pa√≠ses de centroam√©rica
#el .reset_index() resetea el indice 
reporte_por_anio = df_centroamerica.groupby("Year")["Value"].sum().reset_index()

#Renombramos las columnas 
reporte_por_anio.columns = ["Year", "total_por_a√±o"]

#Mostramos el reporte
reporte_por_anio.head()
```

# Pregunta 13

Para s√≥lo los pa√≠ses de Centroam√©rica ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua'] obtener un reporte por a√±o de las emisiones en promedio en Centroam√©rica usando el dataframe 'df_emisions_co2_per_capita'?
Se espera un dataframe de 2 columnas: 'Year' y 'promedio_por_a√±o'

```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta

# Filtramos el DataFrame solo para los pa√≠ses de Centroam√©rica
#el .isin() funciona para filtrar filas que tienen un valor dentro de una lista espec√≠fica
df_centroamerica_per_capita = df_emisions_co2_per_capita[df_emisions_co2_per_capita["Country"].isin(paises_centroamerica)]

# Agrupamos por a√±o y calculamos el promedio de emisiones per c√°pita
#el .mean() calula la media aritmetica
reporte_promedio_por_anio = df_centroamerica_per_capita.groupby("Year")["Value"].mean().reset_index()

# Renombramos columnas
reporte_promedio_por_anio.columns = ["Year", "promedio_por_a√±o"]

# Mostramos resultado
reporte_promedio_por_anio.head()
```

# Pregunta 14

Cu√°l es el promedio de las emisiones per capita para el a√±o 2005 usando el dataframe 'df_emisions_co2_per_capita'?


```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta

# Filtramos solo las filas del a√±o 2005
df_anio_2005 = df_emisions_co2_per_capita[df_emisions_co2_per_capita["Year"] == 2005]

# Calculamos el promedio de la columna Value (emisiones per c√°pita)
promedio_emisiones_per_capita_2005 = df_anio_2005["Value"].mean()

# Mostramos el resultado
print(f"El promedio de emisiones per capita en el a√±o 2005 es de : {promedio_emisiones_per_capita_2005:.0f}")
```

# Pregunta 15

Usa el dataframe  'df_emisions_co2_per_capita' con las columnas Year y promedio_por_a√±o para crear una gr√°fica de l√≠neas que muestre la evoluci√≥n del promedio de emisiones per c√°pita en Centroam√©rica a lo largo del tiempo; el eje X debe ser el a√±o y el eje Y el promedio de emisiones.

```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta

# Creamos la gr√°fica

plt.style.use("ggplot")
plt.rcParams["figure.figsize"] = (15, 5)

# Se crea la figura y los ejes
fig, ax = plt.subplots()

# Variables del gr√°fico
anios = reporte_promedio_por_anio["Year"]
promedios = reporte_promedio_por_anio["promedio_por_a√±o"]

# l√≠nea de emisiones per c√°pita
ax.plot(anios, promedios, marker='o', color='seagreen', label='Emisiones per c√°pita')

# T√≠tulos y etiquetas
ax.set_title('Promedio de emisiones per c√°pita en Centroam√©rica por a√±o', fontsize=16)
ax.set_xlabel('A√±o', fontsize=14)
ax.set_ylabel('Toneladas m√©tricas de CO2 per c√°pita', fontsize=14)
ax.legend()


plt.tight_layout()
plt.show()
```
